```python
#連續爬蟲多頁，例子
import requests
from bs4 import BeautifulSoup

url = "https://www.ptt.cc/bbs/joke/index.html"
for i in range(3): #連續爬3頁
    r = requests.get(url) #將此頁面的HTML給get下來
    soup = BeautifulSoup(r.text, "html.parser") #將網頁資料以html.parser
    sel = soup.select("div.title a") #取html標中的<div class="title"><div>中的<a>標籤，將之存入
    
    u = soup.select("div.btn-group.btn-group-paging a") #上一頁的a標籤from F12
    print("本頁的url為: " + url)
    url = "https://www.ptt.cc" + u[1]["href"] #組合出上一頁網址
    
    for title in sel:
        print(title["href"], title.text) #用迴圈印出網址和標題
```

## 遇到按鈕？ (Ex: 是否年滿18歲?)

```python
#發現按鈕後的封包，以POST方法傳送，且含有cookie和參數
#可藉有requests中的某參數來儲存
r = requests.Session() #此函數會將送出的request所收到的cookies全部存起來，並在發送下一次請求時，送出對應參數
para = {
    'from':'/bbs/Gossiping/index.html',
    'yes':'yes'
} #儲存需發送的參數
#(用firefox打開，按鈕之後，在Network標籤裡面按唯一post的那個請求，然後看params標籤就有)
r1 = r.post("https://www.ptt.cc/ask/over18?from=%2Fbbs%2FGossiping%2Findex.html", data = para) #把參數寫到data參數裡面，然後發post請求
r2 = r.get("https://www.ptt.cc/bbs/Gossiping/index.html") #進去後用一般的get請求
print(r2.text)
```

## 爬取圖片?!

```python
#爬取下列網站的圖
#也是先進到圖片的下載頁面
#先把網址get下來
webpic = requests.get('https://cdn.iconscout.com/icon/premium/png-256-thumb/web-crawler-3-45104.png')
#接著轉為二進制
pic = webpic.content

#接下來在本地端開啟一個「可寫，二進制」的檔案
#寫進去
#FLAG的部分，r是讀取，w是寫入，b是二進制開檔案，a是添加內容
pic_out = open('爬蟲.png', 'wb') #'爬蟲.png'是為圖片檔取的名
pic_out.write(pic) #把get的圖片存進去

#無論之前做了什麼，一定要記得關檔案！非常重要！！！
pic_out.close()
```
