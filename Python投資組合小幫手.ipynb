# Step2: 計算報酬率(profit)/風險(risk)

```python
import tkinter as tk
```

```python
from datetime import *
while True:
    pad=input('請輸入想回測到西元幾年-幾月-幾日(系統只能回朔到2014-6-1):')
    pdd=datetime.strptime(pad,'%Y-%m-%d')
    dpp=datetime.strptime('2014-6-1','%Y-%m-%d')
    td=datetime.today()
    diff1=td-pdd
    diff2=td-dpp
    pddd=int(str(diff1.days))
    dppp=int(str(diff2.days))
    if pddd<dppp:
        print('總共要回測',pddd,'天')         #請使用者輸入pddd
        break
    else:
        print('請輸入2014-6-1以後的日期')
        continue
```

```python
profitA =A.diff()/A     #diff函數是用來計算「差距」
profitA[-pddd:].plot()    #近pddd天的投資報酬率
profitB = B.diff()/B
profitB[-pddd:].plot() 
profitC = C.diff()/C
profitC[-pddd:].plot() 
profitD = D.diff()/D
profitD[-pddd:].plot() 
profitE = E.diff()/E
profitE[-pddd:].plot() 
profitF = F.diff()/F
profitF[-pddd:].plot() 
profitG = G.diff()/G
profitG[-pddd:].plot() 
profitH = H.diff()/H
profitH[-pddd:].plot() 
profitI = I.diff()/I
profitI[-pddd:].plot() 
profitJ = J.diff()/J
profitJ[-pddd:].plot() 

#profit = A.pct_change()       pct_change可以直接算出「百分比」，不用A.diff()/A   可以改成profit= A.pct_change()
#profit[-100:].plot()                 #近pddd天的投資報酬率

```

# 印出「日報酬率」/以及「平均報酬率」（一個數字）

```python
profitA =A.diff()/A   
a=profitA[-pddd:]
print(a)
print(a.mean())

profitB = B.diff()/B
b=profitB[-pddd:] 
print(b)
print(b.mean())

profitC = C.diff()/C
c=profitC[-pddd:]
print(c)
print(c.mean())

profitD = D.diff()/D
d=profitD[-pddd:]
print(d)
print(d.mean())

profitE = E.diff()/E
e=profitE[-pddd:]
print(e)
print(e.mean())

profitF = F.diff()/F
f=profitF[-pddd:]
print(f)
print(f.mean())

profitG = G.diff()/G
g=profitG[-pddd:]
print(g)
print(g.mean())

profitH = H.diff()/H
h=profitH[-pddd:]
print(h)
print(h.mean())

profitI = I.diff()/I
i=profitI[-pddd:]
print(i)
print(i.mean())

profitJ = J.diff()/J
j=profitJ[-pddd:]
print(j)
print(j.mean())
```

# 風險（標準差）（報酬的波動）

```python
# 風險（標準差）（報酬的波動）
riskA = a.std()
riskB = b.std()
riskC = c.std()
riskD = d.std()
riskE = e.std()
riskF = f.std()
riskG = g.std()
riskH = h.std()
riskI = i.std()
riskJ = j.std()
```

# Step4: 夏普指數（供投資人參考，以利未來資產配置）

```python
#下列我們額外乘了一個「252 ** 0.5」
#夏普率= [(每日報酬率平均值- 無風險利率) / (每日報酬的標準差)]x (252平方根)
#作用是算出「年化的夏普指數」(annual sharpe ratio)
#其中的252是一年大約的交易天數
#之所以乘這個常數，是因為乘上這個數字才有辦法跟別人比較！
sharpeA = (a / riskA) * (252 ** 0.5)
sharpeB = (b / riskB) * (252 ** 0.5)
sharpeC = (c / riskC) * (252 ** 0.5)
sharpeD = (d / riskD) * (252 ** 0.5)
sharpeE = (e / riskE) * (252 ** 0.5)
sharpeF = (f / riskF) * (252 ** 0.5)
sharpeG = (g / riskG) * (252 ** 0.5)
sharpeH = (h / riskH) * (252 ** 0.5)
sharpeI = (i / riskI) * (252 ** 0.5)
sharpeJ = (j / riskJ) * (252 ** 0.5)
```

# 畫出夏普指數時序圖

```python
sharpeA.plot()
sharpeB.plot()
sharpeC.plot()
sharpeD.plot()
sharpeE.plot()
sharpeF.plot()
sharpeG.plot()
sharpeH.plot()
sharpeI.plot()
sharpeJ.plot()
```

# 用移動窗格的方式計算移動平均，獲得較為平滑的移動平均線

```python
x = 5 #以「週」為單位
shmuA = sharpeA.rolling(x).mean()
shmuB = sharpeB.rolling(x).mean()
shmuC = sharpeC.rolling(x).mean()
shmuD = sharpeD.rolling(x).mean()
shmuE = sharpeE.rolling(x).mean()
shmuF = sharpeF.rolling(x).mean()
shmuG = sharpeG.rolling(x).mean()
shmuH = sharpeH.rolling(x).mean()
shmuI = sharpeI.rolling(x).mean()
shmuJ = sharpeJ.rolling(x).mean()
```
